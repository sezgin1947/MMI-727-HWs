```
# Importing the Tensorflow library 
import tensorflow as tf 

# Importing the NumPy library 
import numpy as np 

# Importing the matplotlib.pylot function 
import matplotlib.pyplot as plt 

# A vector of size 15 with values from -5 to 5 
a = np.linspace(-15, 15, 30) 

# Applying the sigmoid function and 
# storing the result in 'b' 
b = tf.nn.sigmoid(a) 

# Initiating a Tensorflow session 
with tf.Session() as sess: 
	plt.plot(a, sess.run(b), color = 'red', marker = "x") 
	plt.title("tensorflow.nn.sigmoid") 
	plt.xlabel("X") 
	plt.ylabel("Y") 

	plt.show() 
# Apply ReLu Function and store the result in 'b'
b = tf.nn.relu(a)

# Initiate Tensorflow Session and Plot
with tf.Session() as sess: 
	plt.plot(a, sess.run(b), color = 'red', marker = "x") 
	plt.title("tensorflow.nn.ReLu") 
	plt.xlabel("X") 
	plt.ylabel("Y") 

	plt.show()

  
# Apply Tanh Function and store the result in 'b'
b = tf.nn.tanh(a)

# Initiate Tensorflow Session and Plot
with tf.Session() as sess: 
	plt.plot(a, sess.run(b), color = 'red', marker = "x") 
	plt.title("tensorflow.nn.Tanh") 
	plt.xlabel("X") 
	plt.ylabel("Y") 

	plt.show()
  
# Apply Softmax Function and store the result in 'b'
b = tf.nn.softmax(a)

# Initiate Tensorflow Session and Plot
with tf.Session() as sess: 
	plt.plot(a, sess.run(b), color = 'red', marker = "x") 
	plt.title("tensorflow.nn.Softmax") 
	plt.xlabel("X") 
	plt.ylabel("Y") 

	plt.show()
```
